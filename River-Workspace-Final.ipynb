# The Metric

The metric that I have chosen to show the "best" neighborhood is **the number of social services** in that given neighborhood. My thought process was pretty simple. The more social services in one area, the "better" neighborhood that it is. I am taking data on the Western Pennsylvania Regional Data Center website from a dataset named "BigBurgh Social Service Listings," which is from BigBurgh.com.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('social-service-data.csv', encoding="utf8")

neighborhoods = df.iloc[:, 3]

neighborhood_counts = neighborhoods.value_counts()

sorted_neighborhoods = neighborhood_counts.index

counts = neighborhood_counts.values

plt.figure(figsize=(16, 8))
plt.bar(sorted_neighborhoods, counts)
plt.xlabel('Neighborhoods')
plt.ylabel('Number of Social Services')
plt.title('Social Services in Pittsburgh Neighborhoods')
plt.xticks(rotation=60, ha='right')
plt.yticks(range(max(counts) + 1))
plt.tight_layout()
plt.show()
```

# "The Best Neighborhood"

According to my metric, Downtown Pittsburgh just barely squeaked out a first place finish, with 20 recorded social services. Considering that it is one of the bigger areas in Pittsburgh, it makes sense that it won. Having the most things to do in the city, it would only make sense that it would have the most social services as well. It is important to take this metric with a grain of salt though.
